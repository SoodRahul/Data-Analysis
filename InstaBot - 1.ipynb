#### Importing necessary libraries and modules

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager as CDM
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
```

```python
# used for searching username provided
# needToclick if True then it will implicity click on search icon to enable searching window
def searchUserhandle(username,needToclick=True):
    if needToclick == True:
        searchBox=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[contains(@Class,"x19c4wfv")]')))[1]
        searchBox.click()
    searchBox_input=wait.until(EC.presence_of_element_located((By.XPATH,'//input[@Class="_aauy"]')))
    searchBox_input.send_keys(username)
    return searchBox_input
    
# to click on desired username 
def click_on_searched_username(usernameToSearch):
    checkIfPresent=wait.until(EC.text_to_be_present_in_element((By.XPATH,'//div[contains(@class,"_ab8w  _ab94 _ab97 _ab9f _ab9k _ab9p  _aba0 _aba8 _abcm")]'),usernameToSearch))
    if checkIfPresent is True:
        _Button=wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"_ab8w  _ab94 _ab97 _ab9f _ab9k _ab9p  _aba0 _aba8 _abcm")]')))
        _Button.click()
        wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"_aa_y _aa_z _aa_-")]')))
    else:
        return ("username not present")

# to get the current status if I am following the searched user or not
def getFollowStatus():
    sleep(3)
    soup=BeautifulSoup(driver.page_source)
    status=soup.find('div',{'class':"_aacl _aaco _aacw _aad6 _aade"}).text.strip()
    if status == "Following":
        return ('You Started Following '+usernameToSearch)
    elif status == "Follow":
        return ('You are not Following '+usernameToSearch)


# to like top/first 30 posts of a user
# if like is True then it will like
# if unlike is True then it will unlike 
def get_Top30_posts_liked_and_unliked(like=True,unlike=False):
    
    if like == True:
        status="liked"
    elif unlike == True:
        status = "unliked"
        
    currentScrollHeight=2500
    while True:
        post_link_Arr=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//article[@Class="x1iyjqo2"]//a')))
        if len(post_link_Arr) >= 30:
            break
        driver.execute_script('window.scrollTo(0,arguments[0]);',currentScrollHeight)
        currentScrollHeight+=2500
        
    post_link_Arr=post_link_Arr[:30]

    count=1
    for post in post_link_Arr:
        
        sleep(5)

        post.click()


        like_Button=wait.until(EC.presence_of_element_located((By.XPATH,'//span[@Class="_aamw"]')))
        like_Button.click()

        print(f"Post NO : {count} is "+ status)
        
        post_close_button=driver.find_element_by_xpath('//div[@Class="x10l6tqk x160vmok x1eu8d0j x1vjfegm"]')
        post_close_button.click()
        
        
        count+=1
        

# returrns set of Followers
# you can also pass number of first followers u need to search by setting passNfollowers True  and give value of n
# This is done to avoid TimeOut
def getFollowers_Set(passNfollowers=False,n=0):
    soup=BeautifulSoup(driver.page_source)
    no_of_posts_followers_followingArr=soup.find_all('span',{'class':"_ac2a"})
    if passNfollowers == True:
        nFollowers=n
    
    else:
        nFollowers=int(float(no_of_posts_followers_followingArr[1].text.strip().replace(",","")))

    myProfile_Followers_Button=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//li[contains(@Class," xl565be    x2pgyrj  x1m39q7l x1uw6ca5")]//a')))[0]
    myProfile_Followers_Button.click()

    while True:
        followersArr=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[contains(@Class," _ab8y  _ab94 _ab97 _ab9f _ab9k _ab9p _abcm")]')))
        if len(followersArr) >= nFollowers:
            break
        element=wait.until(EC.presence_of_element_located((By.XPATH,'//span[contains(@Class,"_aap6 _aap7 _aap8")]//a')))
        

        element.send_keys(Keys.END)

    sleep(3)

    soup=BeautifulSoup(driver.page_source)
    followersArr=soup.find_all('span',{'class':"_aacl _aaco _aacw _aacx _aad7 _aade"})


    myFollowersSet=set()
    for i in followersArr:
        myFollowersSet.add(i.text.strip())


    driver.back()
    return myFollowersSet

# Similar to getFollowers_Set but it will return set of following
def getFollowing_Set(passNfollowing=False,n=0):
    
    soup=BeautifulSoup(driver.page_source)
    no_of_posts_followers_followingArr=soup.find_all('span',{'class':"_ac2a"})

    if passNfollowing == True:
        nFollowers=n
    
    else:
        nFollowing=int(float(no_of_posts_followers_followingArr[2].text.strip().replace(",","")))
    
    myProfile_following_button=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//li[contains(@Class," xl565be    x2pgyrj  x1m39q7l x1uw6ca5")]//a')))[-1]
    myProfile_following_button.click()

    while True:
        followingArr=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[contains(@Class," _ab8y  _ab94 _ab97 _ab9f _ab9k _ab9p _abcm")]')))
        if len(followingArr) >= nFollowing:
            break
        element=wait.until(EC.presence_of_element_located((By.XPATH,'//span[contains(@Class,"_aap6 _aap7 _aap8")]//a')))
        

        element.send_keys(Keys.END)

    sleep(5)


    soup=BeautifulSoup(driver.page_source)
    followingArr=soup.find('div',{'class':"_aano"}).div.find_all('span',{'class':"_aacl _aaco _aacw _aacx _aad7 _aade"})
    myfollowingSet=set()
    for i in followingArr:
        myfollowingSet.add(i.text.strip())

    sleep(2)

    driver.back()
    return myfollowingSet
```
